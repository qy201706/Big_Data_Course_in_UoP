## Using Pre-trained Networks for Image Recognition

```python
from IPython.display import Image, display

img_paths = ['/home/kaggle/dog-breed-identification/test/000621fb3cbb32d8935728e48679680e.jpg',
             '/home/kaggle/dog-breed-identification/test/5441b8726acbf1158999122ba0729721.jpg',
             '/home/kaggle/dog-breed-identification/test/e8914f74848c23b70b520887b3ff2e8a.jpg']          

#img_paths = ['/home/kaggle/seefood/test/not_hot_dog/13023.jpg',
#             '/home/kaggle/seefood/test/not_hot_dog/44578.jpg',
#             '/home/kaggle/seefood/test/not_hot_dog/75638.jpg']
                          
display(Image(img_paths [0]))
display(Image(img_paths [1]))
display(Image(img_paths [2]))
```

```python
import numpy as np
from tensorflow.python.keras.applications.resnet50 import preprocess_input
from tensorflow.python.keras.preprocessing.image import load_img, img_to_array

image_size = 224

def read_and_prep_images(img_paths, img_height=image_size, img_width=image_size):
    imgs = [load_img(img_path, target_size=(img_height, img_width)) for img_path in img_paths]
    img_array = np.array([img_to_array(img) for img in imgs])
    output = preprocess_input(img_array)
    return(output)
```

```python
from tensorflow.python.keras.applications import ResNet50

my_model = ResNet50(weights='/home/kaggle/resnet50/resnet50_weights_tf_dim_ordering_tf_kernels.h5')

test_data = read_and_prep_images(img_paths)
preds = my_model.predict(test_data)
```

```python
from tensorflow.python.keras.applications.imagenet_utils import decode_predictions
from IPython.display import Image, display

#most_likely_labels = decode_predictions(preds, top=3, class_list_path='/home/kaggle/resnet50/imagenet_class_index.json')
most_likely_labels = decode_predictions(preds, top=3)

for i, img_path in enumerate(img_paths):
  display(Image(img_paths[i]))
  print(most_likely_labels[i])
```

## Transfer Learning

```python
from tensorflow.python.keras.applications import ResNet50
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Flatten, GlobalAveragePooling2D

# num_classes is the number of categories your model chooses between for each prediction
num_classes = 2
resnet_weights_path = '/home/kaggle/resnet50/resnet50_weights_tf_dim_ordering_tf_kernels_notop.h5'
    # Note this is *different* to weights file above.

my_new_model = Sequential()
my_new_model.add(ResNet50(include_top=False, pooling='avg', weights=resnet_weights_path))
my_new_model.add(Dense(num_classes, activation='softmax'))

# We don't want to train the first layers - ResNet50 is already trained
my_new_model.layers[0].trainable = False
```

```python
my_new_model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.python.keras.applications.resnet50 import preprocess_input
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator

image_size = 224
data_generator = ImageDataGenerator(preprocessing_function=preprocess_input)

train_generator = data_generator.flow_from_directory(
        '/home/kaggle/rural_and_urban_photos/train',
        target_size=(image_size, image_size),
        batch_size=24,
        class_mode='categorical')

validation_generator = data_generator.flow_from_directory(
        '/home/kaggle/rural_and_urban_photos/val',
        target_size=(image_size, image_size),
        class_mode='categorical')

my_new_model.fit_generator(
        train_generator,
        steps_per_epoch=3,
        validation_data=validation_generator,
        validation_steps=1)
```

```python
from os import listdir
from os.path import join, isdir

image_dir = '/home/kaggle/rural_and_urban_photos/val'
classes = listdir(image_dir)
class_dirs = [join(image_dir, dirname) for dirname in classes if isdir(join(image_dir, dirname))]
img_paths = [join(class_dir, filename) for class_dir in class_dirs for filename in listdir(class_dir)]

# uncomment following if you want to be convinced the Python os code above is finding the right paths.
# print(img_paths)

test_data = read_and_prep_images(img_paths)
preds = my_new_model.predict(test_data)
class_indices = validation_generator.class_indices  # see documentation for ImageDataGenerator class referenced above
print(class_indices)

for i, img_path in enumerate(img_paths):
    display(Image(img_path))
    print(preds[i])
```

## Data Augmentation

```python
from tensorflow.python.keras.applications.resnet50 import preprocess_input
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator

image_size = 224

data_generator_with_aug = ImageDataGenerator(preprocessing_function=preprocess_input,
                                   horizontal_flip=True,
                                   width_shift_range = 0.2,
                                   height_shift_range = 0.2)

train_generator = data_generator_with_aug.flow_from_directory(
        '/home/kaggle/rural_and_urban_photos/train',
        target_size=(image_size, image_size),
        batch_size=24,
        class_mode='categorical')

data_generator_no_aug = ImageDataGenerator(preprocessing_function=preprocess_input)
validation_generator = data_generator_no_aug.flow_from_directory(
        '/home/kaggle/rural_and_urban_photos/val',
        target_size=(image_size, image_size),
        class_mode='categorical')

my_new_model.fit_generator(
        train_generator,
        steps_per_epoch=3,
        epochs=2,
        validation_data=validation_generator,
        validation_steps=1)
```

```python
preds = my_new_model.predict(test_data)
class_indices = validation_generator.class_indices  # see documentation for ImageDataGenerator class referenced above
print(class_indices)

for i, img_path in enumerate(img_paths):
    display(Image(img_path))
    print(preds[i])
```
